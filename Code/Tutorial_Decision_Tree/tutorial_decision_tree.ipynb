# Mô hình câu quyêt dịnh - Decision tree

mô hình học tổ hợp

```python
from sklearn import datasets

iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target # giá trị đầu ra
```

```python
X # ma trận - mỗi hàng 1 đặc trưng của mẫu dữ liệu 
# ==> 4 đặc trưng ==> tổng số hàng = 150 mẫu, mỗi mẫu có 4 đặc trưng
```

```python
X.shape
```

```python
y # 3 [0, 1, 2] classes of iris species names
```

```python
y.shape
```

# Build model

## Dùng decision tree

```python
from sklearn.tree import DecisionTreeClassifier # DecisionTreeRegressor --> hồi quy

model = DecisionTreeClassifier() 
model.fit(X, y)
```

# Trưc quan hóa mô hình

```python
from sklearn import tree

presentation = tree.export_text(model)
print(presentation) # in ra cây quyết định - giống cây thư mục trong window


# tên feature thứ i tương ứng với cột thứ i trong X
# ==> Rất khó theo dõi 
# khó hình dung do không biết feature, class này là gì 
```

# Dùng matplotlib đê trục quan dê hơn

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(25, 20))
_ =  tree.plot_tree(model,
                    feature_names=iris.feature_names,
                    class_names=iris.target_names, # bình thường là 0,1,2 nhưng giờ gán tên vào
                    filled=True)
plt.show() # vẽ cây quyết định với tên feature và tên class
#hệ số gini từ 0->1 đo hỗn tạp 
#giá trị đầu đk kiểm tra, thỏa thì đi bên trái, không thỏa thì đi bên phải
# samples là số mẫu thỏa cái đó
# values là có thể vào mẫu nào thuộc lớp nào
# ta chỉ đưa ra kết luận khi gini = 0

```

- Hệ số gini - độ thuần khiết của lớp [0, 1] || đo độ hỗn tạp   
- Cấu trúc cây  dễ bị overfitting - vì nó quá phức tạp
- ==> Tránh overfitting:   
    * pruning - cắt tỉa cây quyết định (nhánh có: 1. hệ số gini quá tháp, 2: nhánh quá tiểu tiết) ==> giảm độ phức tạp của cây quyết định
    * max_depth - Truyền thêm thám số ==> giới hạn chiều sâu của cây quyết định ==> Giới hạn độ sâu

# Đánh giá mô hình

```python
y_pred = model.predict(X)
print("Accuracy: ", sum(y_pred == y) / len(y))


```

```python
from sklearn.metrics import accuracy_score
y_pred = model.predict(X)
acc = accuracy_score(y, y_pred)
print("Accuracy: ", acc)
```
